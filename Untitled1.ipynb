```python
#1 (insert)
n = 4
arr = [10, 20, 30, 40, 0]
ele = 50
print("Array before insertion")
for i in range(n):
    print(arr[i], end=" ")
    arr[n] = ele

print("\nArray after insertion")
for i in range(n + 1):
    print(arr[i], end=" ")

#2 (Largest element)
def largest(arr, n):
    mx = arr[0]
    for i in range(1, n):
        if arr[i] > mx:
            mx = arr[i]
    return mx
arr = [10, 324, 45, 90, 9808]
n = len(arr)
ans = largest(arr, n)
print(ans)

#3 (2nd largest)
def getSecondLargest(arr):
    n = len(arr)
    arr.sort()
    for i in range(n - 2, -1, -1):
        if arr[i] != arr[n - 1]:
            return arr[i]
    return -1
arr = [12, 35, 1, 10, 34, 1]
print(getSecondLargest(arr))

#4 (Zeros)
def pushZerosToEnd(arr):
    n = len(arr)
    temp = [0] * n
    j = 0
    for i in range(n):
        if arr[i] != 0:
            temp[j] = arr[i]
            j += 1
    while j < n:
        temp[j] = 0
        j += 1
    for i in range(n):
        arr[i] = temp[i]
arr = [1, 2, 0, 4, 3, 0, 5, 0]
pushZerosToEnd(arr)
for num in arr:
    print(num, end=" ")

#5 (Rotate)
arr=[1,2,3,4,5]
last=arr[-1]
for i in range(len(arr)-1,0,-1):
    arr[i]=arr[i-1]
arr[0]=last
for i in range(len(arr)):
    print(arr[i])

#6 (check if sorted)
arr=[1,2,3,4,5,6]
n=len(arr)
for i in range(n):
  if arr[i-1]>arr[i]:
    print("Not Sorted")
    break
  print("Sorted")

#7 (Reverse a string)
a=input("Enter a string : ")
print("Reversed string : ", a[::-1])

#8 (Palindrome)
a=input("Enter a string : ")
if a==a[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

#9 (count)
arr=[1,2,3,2,3,1,4,2,4]
freq={}
for i in arr:
    if i in freq:
        freq[i]+=1
    else:
        freq[i]=1
for i in freq:
  print(i, "occurs", freq[i], "times")

#10 (Reverse an array)
a=[1,2,3,4,5]
x=a[::-1]
print("Reversed array : ", x)









```
